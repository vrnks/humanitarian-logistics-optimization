```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

file = pd.read_excel("HRPR Weekly Digest Bulletin.xlsx")
file
```

```python
file = file.drop(index=0)
file
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
file.dtypes
```

```python
file.shape
```

```python
# Описова статистика за допомогою методу describe 
stats = file['Quantity'].describe()
stats
```

```python

plt.rcParams["font.family"] = "Times New Roman"
plt.figure(figsize=(12, 8))
file.groupby('Oblast / Region')['Quantity'].sum().plot(kind='bar', color='blue')
plt.title('Кількість продуктів, в яких є потреба по областях')
plt.xlabel('Область')
plt.ylabel('Кількість продуктів')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
oblasts = [
    'Donetska', 'Zaporizka', 
    'Mykolaivska', 'Odeska', 'Sumska', 'Zakarpatska', 'Volynska', 'Khmelnytska', 'Ivano-Frankivska', 'Ternopilska', 'Chernihivska', 'Cherkaska',  'Kharkivska', 'Khersonska','Kirovohradska',  'Lvivska', 'Kyiv',
]

data = file[file['Oblast / Region'].isin(oblasts)]
data.head()
```

```python
# Створіть описову статистику за допомогою методу describe 
stats = data['Quantity'].describe()
stats
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.figure(figsize=(12, 8))
data.groupby('Oblast / Region')['Quantity'].sum().plot(kind='bar', color='blue')
plt.title('Кількість продуктів, в якій є потреба по областях (без Дніпропетровської)')
plt.xlabel('Область')
plt.ylabel('Кількість продуктів')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
