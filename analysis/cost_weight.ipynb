```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

file = pd.read_excel("HRPR Weekly Digest Bulletin.xlsx")
file
```

```python
file = file.drop(index=0)
file
```

```python
file.shape
```

```python
total_price_by_region = file.groupby('Oblast / Region')['price total'].sum()
total_weight_by_region = file.groupby('Oblast / Region')['weight total'].sum()
total_weight_by_region = total_weight_by_region/1000000

print(total_weight_by_region)
print(total_price_by_region)
```

```python

plt.rcParams["font.family"] = "Times New Roman"
# Побудова стовпчикової діаграми для загальної ціни по областях
plt.figure(figsize=(12, 8))
total_price_by_region.plot(kind='bar', color='blue')
plt.title('Загальна ціна по областях')
plt.xlabel('Область')
plt.ylabel('Загальна ціна, грн')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python

plt.figure(figsize=(12, 8))
total_weight_by_region.plot(kind='bar', color='green')
plt.title('Загальна вага по областях')
plt.xlabel('Область')
plt.ylabel('Загальна вага (т)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
